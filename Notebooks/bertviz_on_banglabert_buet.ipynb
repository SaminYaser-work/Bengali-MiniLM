# ট্র্যান্সফরমারের কার্যপ্রণালীয়ের চিত্র স্ফুটন

```python
from bertviz import head_view, model_view
from transformers import BertTokenizer, BertModel

sentence_a = "আজ আমার মন ভালো নেই"
sentence_b = "আমার ব্র্যাকে ভর্তি হওয়া উচিত ছিল"

model_version = 'csebuetnlp/banglabert'
model = BertModel.from_pretrained(model_version, output_attentions=True)
tokenizer = BertTokenizer.from_pretrained(model_version)
inputs = tokenizer.encode_plus(sentence_a, sentence_b, return_tensors='pt')
input_ids = inputs['input_ids']
token_type_ids = inputs['token_type_ids']
attention = model(input_ids, token_type_ids=token_type_ids)[-1]
sentence_b_start = token_type_ids[0].tolist().index(1)
input_id_list = input_ids[0].tolist() # Batch index 0
tokens = tokenizer.convert_ids_to_tokens(input_id_list) 
```

# ট্র্যান্সফরমারের মাথার দৃশ্য

এখানে ট্র্যান্সফরমারের একটি মাত্র লেয়ারের মাথাসমূহের মনোযোগ দৃশ্যপটে তুলে ধরা হয়েছে। প্রত্যেকটি সরল রেখা বাম দিকের একটি টোকেন এর সাথে ডান দিকের একটি টোকেন মধ্যে থাকা মনোযোগকে নির্দেশ করে (যার মান ০ থেকে ১ এর মধ্যে) এবং সরল রেখার রঙ এক একটি মাথাকে নির্দেশ করে। একাধিক রং দ্বারা একাধিক মাথাকে বোঝানো হয়েছে।

```python
head_view(attention, tokens, sentence_b_start)
```
